```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 준비
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
data = np.random.randn(1000)

plt.figure(figsize=(6, 4))

# 전체 제목
plt.suptitle("plt.subplot()으로 그린 2행 3열 그래프 예제", fontsize=16)

# (1,1): 선 그래프
plt.subplot(2, 3, 1)
plt.plot(x, y, "b-")
plt.title("Sine Wave")

# (1,2): 코사인 그래프
plt.subplot(2, 3, 2)
plt.plot(x, z, "r-")
plt.title("Cosine Wave")

# (1,3): 산점도
plt.subplot(2, 3, 3)
plt.scatter(x, y, c="g", s=5)
plt.title("Scatter Plot")

# (2,1): 히스토그램
plt.subplot(2, 3, 4)
plt.hist(data, bins=30, color="purple")
plt.title("Histogram")

# (2,2): 박스플롯
plt.subplot(2, 3, 5)
plt.boxplot(data)
plt.title("Box Plot")

# (2,3): 막대그래프
plt.subplot(2, 3, 6)
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]
plt.bar(categories, values, color='orange')
plt.title("Bar Chart")

# 레이아웃 조정
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# x 범위 (0,1) 사이
x = np.linspace(0.01, 0.99, 400)
y1 = np.log(x)
y2 = np.log(1 - x)

plt.figure(figsize=(6, 3))

# log(x)
plt.subplot(1, 2, 1)
plt.plot(x, y1, label="log(x)")
plt.ylim(-5, 0.5)  # y축 동일하게
plt.title("log(x)")
plt.grid(True)
plt.legend()

# log(1-x)
plt.subplot(1, 2, 2)
plt.plot(x, y2, label="log(1-x)")
plt.ylim(-5, 0.5)  # y축 동일하게
plt.title("log(1-x)")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# x 값 정의
x = np.linspace(-5, 5, 400)

# 함수 정의
y1 = 3 * x + 1               # y = 3x + 1
y2 = x**2                    # y = x^2
y3 = np.exp(x)               # y = e^x
y4 = np.maximum(0, x)        # ReLU
y5 = 1 / (1 + np.exp(-x))    # Sigmoid
y6 = np.tanh(x)              # tanh

# subplot: 2행 3열
fig = plt.figure(figsize=(6, 5))

# 1
plt.subplot(2, 3, 1)
plt.plot(x, y1, label="y = 3x + 1")
plt.legend()
plt.grid(True)

# 2
plt.subplot(2, 3, 2)
plt.plot(x, y2, label="y = x^2")
plt.legend()
plt.grid(True)

# 3
plt.subplot(2, 3, 3)
plt.plot(x, y3, label="y = e^x")
plt.ylim(-1, 50)  # 보기 좋게 제한
plt.legend()
plt.grid(True)

# 4
plt.subplot(2, 3, 4)
plt.plot(x, y4, label="ReLU")
plt.legend()
plt.grid(True)

# 5
plt.subplot(2, 3, 5)
plt.plot(x, y5, label="Sigmoid")
plt.legend()
plt.grid(True)

# 6
plt.subplot(2, 3, 6)
plt.plot(x, y6, label="tanh")
plt.legend()
plt.grid(True)

fig.suptitle("다양한 그래프", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 준비
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)

# 2행 3열 subplot 생성
fig, axes = plt.subplots(2, 3, figsize=(7, 4))

# 전체 제목 추가
fig.suptitle("다양한 그래프 예제 (2행 3열 Subplots)", fontsize=16)

# (0, 0): 선 그래프
axes[0, 0].plot(x, y, color='b')
axes[0, 0].set_title("Sine Wave")

# (0, 1): 코사인 곡선
axes[0, 1].plot(x, z, color='r')
axes[0, 1].set_title("Cosine Wave")

# (0, 2): 산점도
axes[0, 2].scatter(x, y, color='g')
axes[0, 2].set_title("Scatter Plot")

# (1, 0): 히스토그램
data = np.random.randn(1000)
axes[1, 0].hist(data, bins=30, color='purple')
axes[1, 0].set_title("Histogram")

# (1, 1): 박스플롯
axes[1, 1].boxplot(data)
axes[1, 1].set_title("Box Plot")

# (1, 2): 막대그래프
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]
axes[1, 2].bar(categories, values, color='orange')
axes[1, 2].set_title("Bar Chart")

#공통 x,y축 라벨
fig.supxlabel("공통 x축 라벨", fontsize =12)
fig.supylabel("공통 y축 라벨", fontsize =12)

# 전체 제목 공간 확보
plt.tight_layout()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.01, 0.99, 400)
y1 = np.log(x)
y2 = np.log(1 - x)

fig, axes = plt.subplots(1, 2, figsize=(7, 3), sharey=True)

axes[0].plot(x, y1, label="log(x)")
axes[0].set_title("log(x)")
axes[0].grid(True)
axes[0].legend()

axes[1].plot(x, y2, label="log(1-x)")
axes[1].set_title("log(1-x)")
axes[1].grid(True)
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# x 값 정의
x = np.linspace(-5, 5, 400)

# 함수 정의의 리스트
functions = [
    (3*x + 1, "y = 3x + 1"),
    (x**2, "y = x^2"),
    (np.exp(x), "y = e^x"),
    (np.maximum(0, x), "ReLU"),
    (1 / (1 + np.exp(-x)), "Sigmoid"),
    (np.tanh(x), "tanh")
]

# 2행 3열 subplots 생성
fig, axes = plt.subplots(2, 3, figsize=(7, 6))

# flatten으로 1차원 배열로 변환
axes = axes.flatten()

# 각 subplot에 함수 그리기
for ax, (y, label) in zip(axes, functions):
    ax.plot(x, y, label=label)
    ax.set_title(label)
    ax.grid(True)
    ax.legend()

    if label == "y = e^x":
        ax.set_ylim(-1, 50)  # 지수함수는 제한 필요

fig.suptitle("다양한 그래프", fontsize=16)

plt.tight_layout()
plt.show()
```

###함수 add_subplot()

```python
import matplotlib.pyplot as plt

fig = plt.figure()

# 2행 2열 중 첫 번째 위치
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot([1, 2, 3], [1, 2, 3])
ax1.set_title("선 그래프")

# 2행 2열 중 두 번째 위치
ax2 = fig.add_subplot(2, 2, 2)
ax2.bar([1, 2, 3], [3, 2, 1])
ax2.set_title("막대 그래프")

# 2행 2열 중 세 번째 위치
ax3 = fig.add_subplot(2, 2, 3)

# 2행 2열 중 네 번째 위치
ax4 = fig.add_subplot(2, 2, 4)

fig.suptitle("선과 막대 그래프")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# x 값 정의
x = np.linspace(-5, 5, 400)

# 함수 정의 리스트
functions = [
    (3*x + 1, "y = 3x + 1"),
    (x**2, "y = x^2"),
    (np.exp(x), "y = e^x"),
    (np.maximum(0, x), "ReLU"),
    (1 / (1 + np.exp(-x)), "Sigmoid"),
    (np.tanh(x), "tanh")
]

# Figure 생성
fig = plt.figure(figsize=(8, 6))

# add_subplot()으로 2행 3열 배치
for i, (y, label) in enumerate(functions, 1):
    ax = fig.add_subplot(2, 3, i)  # 행, 열, 위치
    ax.plot(x, y, label=label)
    ax.set_title(label)
    ax.grid(True)
    ax.legend()

    if label == "y = e^x":
        ax.set_ylim(-1, 50)  # 지수함수 범위 제한

fig.suptitle("다양한 그래프", fontsize=16)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 준비
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
data = np.random.randn(1000)

# Figure 생성
fig = plt.figure(figsize=(8, 4))
fig.suptitle("fig.add_subplot()을 이용한 2행 3열 그래프 예제", fontsize=16)

# (1, 1): Sine Wave
ax1 = fig.add_subplot(2, 3, 1)
ax1.plot(x, y, color='b')
ax1.set_title("Sine Wave")

# (1, 2): Cosine Wave
ax2 = fig.add_subplot(2, 3, 2)
ax2.plot(x, z, color='r')
ax2.set_title("Cosine Wave")

# (1, 3): Scatter Plot
ax3 = fig.add_subplot(2, 3, 3)
ax3.scatter(x, y, color='g')
ax3.set_title("Scatter Plot")

# (2, 1): Histogram
ax4 = fig.add_subplot(2, 3, 4)
ax4.hist(data, bins=30, color='purple')
ax4.set_title("Histogram")

# (2, 2): Box Plot
ax5 = fig.add_subplot(2, 3, 5)
ax5.boxplot(data)
ax5.set_title("Box Plot")

# (2, 3): Bar Chart
ax6 = fig.add_subplot(2, 3, 6)
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]
ax6.bar(categories, values, color='orange')
ax6.set_title("Bar Chart")

# 레이아웃 조정
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
