### 산점도, 논리

```python
#산점
import numpy as np
import matplotlib.pyplot as plt

# Fixing random state for reproducibility
np.random.seed(2025)

N = 80
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (20 * np.random.rand(N)) ** 2

plt.scatter(x, y, s=area, c=colors, alpha=0.7)
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(-100, 100, 1000)  # 1000개의 랜덤 값 추출
y = np.random.randint(-100, 100, 1000)  # 1000개의 랜덤 값 추출

mask1 = abs(x) > 50   # x에 저장된 값 중 절댓값이 50보다 큰 값 걸러냄
mask2 = abs(y) > 50   # y에 저장된 값 중 절댓값이 50보다 큰 값 걸러냄

x = x[mask1 | mask2]  # mask1과 mask2 중 하나라도 만족하는 값 저장
y = y[mask1 | mask2]

plt.scatter(x, y, c=x, cmap='jet', alpha=0.4)
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(-100, 100, 1000)  # 1000개의 랜덤 값 추출
y = np.random.randint(-100, 100, 1000)  # 1000개의 랜덤 값 추출

mask1 = abs(x) > 50   # x에 저장된 값 중 절댓값이 50보다 큰 값 걸러냄
mask2 = abs(y) > 50   # y에 저장된 값 중 절댓값이 50보다 큰 값 걸러냄

x = x[mask1 * mask2]  # mask1과 mask2 모두 만족하는 값 저장
y = y[mask1 * mask2]

# 아래처럼 써도 같음
# x = x[mask1 & mask2]
# y = y[mask1 & mask2]

plt.scatter(x, y, c=x, cmap='jet', alpha=0.5)
plt.colorbar()
plt.show()
```

##교수님하고 실습

```python
import matplotlib.pyplot as plt #subplot

# 전체 제목 - 어느 위치에서 호출해도 무관 (Figure 레벨)
plt.suptitle('전체 제목', fontsize=14, y=1.02)

# subplot 1
plt.subplot(2, 3, 1)
plt.plot([1, 4, 9, 16])
plt.title('제곱수')

# subplot 5
plt.subplot(2, 3, 5)
plt.plot([1, 2, 3, 4])
plt.title('선형')

# subplot 2
plt.subplot(2, 3, 2)
plt.plot([6, 5, 2, 1])
plt.title('감소')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt #subplots

fig, ax = plt.subplots(2, 3, figsize=(7, 4))
fig.set_size_inches(6, 3)

ax[0, 1].plot([1, 4, 9, 16])
ax[1, 0].plot([1, 2, 3, 4])
ax[1, 2].plot([6, 5, 2, 1])

# 전체 제목 - Figure 레벨
fig.suptitle('전체 제목', fontsize=14, fontweight='bold')

# 각 subplot 제목 - Axes 레벨
ax[0, 0].set_title('(0,0) 빈 플롯')
ax[0, 1].set_title('(0,1) 제곱수')
ax[0, 2].set_title('(0,2) 빈 플롯')
ax[1, 0].set_title('(1,0) 선형')
ax[1, 1].set_title('(1,1) 빈 플롯')
ax[1, 2].set_title('(1,2) 감소')

plt.tight_layout()
plt.show()
```

```python
#불리안
import pandas as pd

s = pd.Series([3, 6, -3, 0, -4, 8, -7])
print(s)
print()

print(s > 0)
print()

print(s % 2 == 0)
print()

print(s[s > 0])
print()

print(s[~(s > 0)])
print()

print(s[s % 2 == 0])
print()

print(s[s % 2 != 0])
print()

print(s[(s > 0) & (s % 2 == 0)])
print()
```

```python
#시리즈 정렬
import numpy as np
import pandas as pd

s = ['희빈', '수영', '현수', '지호', '지민']
p = pd.Series([99, 84, 92, 65, 78], index=s)
print(p)
print()

p.name = '수학성적'
p.index.name = 'st_name'
print(p)
print()

print(p.sort_values())
print()

print(p.sort_values(ascending=False))
print()

print(p.sort_index())
print()

print(p.sort_index(ascending=False))
print()
```

```python
#시리즈의 기본 통계 함수
import numpy as np
import pandas as pd

s = ['희빈', '수영', '현수', '지호', '지민']
p = pd.Series([99, 84, 92, 65, 78], index=s)
p.name = '수학성적'

print(p)
print()

print(p.max())
print(p.min())
print()

print(p.mean())
print(p.median())
print()

print(p.count())
print(p.size)
print()

print(p.describe())
print()
```

```python
#메소드 count(), value_counts()
import numpy as np
import pandas as pd

s = pd.Series([3, 1, 1, 2, None, np.nan, 4, 6, 6, np.nan])
print(s)
print()

print(s.values)
print()

print(s.count())# 결측값 제외함
print()

print(s.value_counts()) #고유한 값의 개수를 포함하는 시리즈를 반환 많은 값 으로 내림차순
print()

print(s.value_counts(dropna=False)) #결측값도 포함
print()
```

```python
#s.value_counts(bins=4)
import numpy as np
import pandas as pd

s = pd.Series([3, 1, 1, 2, None, np.nan, 4, 6, 6, np.nan])
print(s)
print()

print(s.value_counts(sort=False, dropna=False))
print()

print(s.value_counts(bins=4)) #시험에는 안 나옴
print()

print(s.value_counts(bins=4, sort=False))
print()

print(s.value_counts(bins=[0.5, 3.5, 6.5], sort=False))
print()
```
