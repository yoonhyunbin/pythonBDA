###Vstack()

```python
import numpy as np

a = np.arange(6).reshape(3, 2)
print(a)

print()

b = np.arange(6, 12).reshape(3, 2)
print(b)

print()

print(np.vstack((a, b)))

x = np.array([1, 2, 3])
print(x)

print()

y = np.array([5, 6, 7])
print(y)

print()

print(np.vstack((x, y)))
```

###Hstack()

```python
import numpy as np #hstack 1차원 1차원 합치면 2차원 vstack 1차원 1차원 합치면 2차원

a = np.arange(6).reshape(3, 2)
print(a)

print()  # 공백 한 줄

b = np.arange(10, 19).reshape(3, 3)
print(b)

print()  # 공백 한 줄

print(np.hstack((a, b)))

print()  # 공백 한 줄

x = np.arange(3)
print(x)

print()  # 공백 한 줄

y = np.arange(10, 15)
print(y)

print()  # 공백 한 줄

print(np.hstack((x, y)))
```

### Column_stack


```python
import numpy as np

x = np.array([1, 2, 3])
print(x)

print()  # 공백 한 줄

y = np.array([10, 20, 30])
print(y)

print()  # 공백 한 줄

print(np.column_stack((x, y)))

print()  # 공백 한 줄

print(np.column_stack((x[:, np.newaxis], y[:, np.newaxis])))

print()  # 공백 한 줄

print(x[:, np.newaxis])

print()  # 공백 한 줄

print(np.hstack((x, y)))

print()  # 공백 한 줄

print(np.column_stack == np.hstack)
```

###concatennate()

```python
#1차원
import numpy as np

a = np.array([1, 2])
print(a)

print()

b = np.array([3, 4])
print(b)

print()

c = np.array([5, 6])
print(c)

print()

print(np.concatenate((a, b, c)))

print()
print(np.concatenate((a, b, c), axis=0))

print()

print(np.concatenate((a, b, c), axis=None))

#2차원 concatennate() =  axis값을 안 넣으면 0값임 이게 0값이면 vstack처럼 나오고 값이 1이면 hstack결과가 나옴 None는 평탄화 hstack하고 비슷한거 같지만 다름 2차원을 1차원으로 hstack처럼 해버릴수 있
x = np.arange(6).reshape(2, 3)
print(x)

print()

y = np.arange(10, 16).reshape(2, 3)
print(y)

print()

z = np.arange(20, 24).reshape(2, 2)
print(z)

print()

print(np.concatenate((x, y)))

x = np.arange(6).reshape(2, 3)
print(x)

print()

y = np.arange(10, 16).reshape(2, 3)
print(y)

print()

z = np.arange(20, 24).reshape(2, 2)
print(z)

print()

print(np.concatenate((x, y), axis=1))

print()

print(np.concatenate((x, y, z), axis=1))

print()

print(np.concatenate((x, y), axis=None))

x = np.arange(6).reshape(2, 3)
print(x)

print()

y = np.arange(10, 16).reshape(2, 3)
print(y)

print()
z = np.arange(20, 24).reshape(2, 2)
print(z)

print()

print(np.concatenate((x, y), axis=1))

print()

print(np.concatenate((x, y, z), axis=1))

print()

print(np.concatenate((x, y), axis=None))
```

###np.r_[] 개요
