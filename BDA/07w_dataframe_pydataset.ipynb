##교수님과 실습

```python
import pandas as pd
import numpy as np

df = pd.DataFrame([[np.nan, 2, np.nan, 0],
                   [3, 4, np.nan, 1],
                   [np.nan, np.nan, np.nan, np.nan],
                   [np.nan, 3, np.nan, 4]],
                  columns = list("ABCD"))
df
```

```python
df.fillna(10)
```

```python
df
```

```python
df.fillna("missing", axis = 0)
```

```python
df.fillna("missing", limit=1)
```

```python
df.fillna("missing", limit=1, axis = 1)
```

```python
df.fillna("missing", limit=2, axis = 1)
```

```python
s = pd.Series([100, 200, 400], index=list("ACD"))
s
```

```python
df
```

```python
df.fillna(value = s, limit=1)
```

```python
value = {"A": 100, "B": 200, "C": 300, "D": 400}
df.fillna(value = value)
```

```python
value = {"A": 100, "B": 200, "C": 300, "D": 400}
df.fillna(value = value, limit=2)
```

```python
df2 = pd.DataFrame(-np.ones((4,4)), columns = list("ABCD"))
df2
```

```python
df.fillna(df2)
```

```python
df.fillna(df2, axis = 1)
```

```python
df
```

```python
df.fillna(method = "ffill")
```

```python
df.fillna(method = "bfill")
```

```python
df.fillna(method = "ffill", limit = 1)
```

```python
df.fillna(df.mean())
```

```python
df.fillna(df.mean()["A":"B"])
```

```python
df.fillna(df.mean()["D"])
```

```python
df.fillna(df.mean()["D"], limit = 1)
```

```python
df.fillna(df.mean()["D"], limit = 1, axis = 1)
```
