##중요
ndarray.shape :배열의차원을나타내는튜플.각차원의크기를표현<br>
ndarray.ndim  :배열의차원수<br>
ndarray.size  :배열의총요소수<br>
ndarray.dtype :배열요소의데이터형식<br>
ndarray.itemsize :배열의각요소의크기(바이트)<br>
ndarray.nbytes   :배열의전체크기(바이트)<br>
ndarray.data     :배열요소를실제로포함하고있는버퍼의시작위치<br>

```python
import numpy as np #ndarry 여러 개의 값을 한 번에 저장하는 자료형

a = np.ndarray((2, 3))
print(a
      )
a = np.ndarray((3, 2), dtype = int)
print(a)

a = np.ndarray((3, 2), int)
print(a)
```

```python
import numpy as np
b = np.array([1, 2, 3], dtype=int)
b

type(b)

print(b.shape)
print(b.ndim)
print(b.dtype)

print(b.size)
print(b.itemsize)
print(b.nbytes)
```

###linspace


```python
#start = 시작 값, stop = 끝 값, num =  생성할 점의 개수 (start-stop) / (num-1) = 0, 10, 6 = 10/5 = 2; 점의 간격 2
import numpy as np
print(np.linspace(2.0, 3.0)) #기본 num은 50개
print((3-2) / (50 - 1))
print(np.linspace(2.0, 3.0, retstep = True)) #num값을 보여줌
a = (np.linspace(2.0, 3.0, retstep = True))
print(a[1])
```

```python
np.linspace(2.0, 3.0, endpoint=False, retstep = True) #endpoint 마지막값 stop값 삭제
```

### 2차원 배열 슬라이싱

### 배열 관련 함수들

```python
import numpy as np

a = np.arange(12).reshape(2, 3, 2)
print(a)

b = a.ravel()
b
```

```python
import numpy as np

a = np.arange(1, 7).reshape(2, 3)
print(a)
print()
print(np.transpose(a)) #행과 열을 바꿔버림

print(a.transpose())

print(a.T)
```

```python
import numpy as np

b = np.arange(1, 13).reshape(3, 2, 2)
print(b)
print(".......................")
print(b.T) #3차원은 (0,0,1) = (1,0,0) and (1,1,0) = (0,1,1) 로 바꿈
```

```python
import numpy as np

x = np.array([[7, 8, 9]])
print(x)
print()

y = np.swapaxes(x, 0, 1)
print(y)
print()

y = np.swapaxes(x, 1, 0)
print(y)
print()

y[0, 0] = 10
print(x)
```

```python
import numpy as np

x = np.array(np.arange(8).reshape(2, 2, 2))
print(x)
print()

print(np.swapaxes(x, 0, 2))
print()

print(np.transpose(x, (2, 1, 0)))
```

```python
import numpy as np

org = np.arange(5)
print(org)
print()

v = org.view()
print(v)
print()

v[2] = 200
print(v)
print()

print(org)
```

```python
import numpy as np

org = np.arange(5)
print(org)
print()

c = org.copy()
print(c)
print()

c[2] = 200
print(c)
print()

print(org)

print(org.base) #배열이 뷰랑 복사본인지 확인 가능 none값 출력
```

##브로드캐스팅

```python
from numpy import arange #열과 열을 연산하는데 개수가 틀리면 진행x 행 과 열은 노 상관

x = arange(1, 10).reshape(3, 3)
print(x)

y = arange(-1, 2)
print(y)

print(x * y)

print(x + y)
```

```python
import numpy as np
from numpy import arange

x = arange(1, 10).reshape(3, 3)
print(x)

y = arange(3, 10, 3).reshape(3, 1)
print(y)

# 소수점 이하 자릿수를 2자리로 설정
np.set_printoptions(precision=2)

print(x / y)

print(x - y)
```

```python
import numpy as np
from numpy import arange

x = arange(1, 4)
print(x)

y = arange(1, 4).reshape(3, 1)
print(y)

print(x * y)

print(x / y)
```

```python
import numpy as np
from numpy import array

x = array([0, 10, 20, 30])
print(x)

print(np.newaxis is None)

# 차수를 증가 1차원을 2차원으로
a = x[None, :]
print(a)

a = x[np.newaxis, :] #열 진행
print(a)

a = x[:, np.newaxis] #행 진행
print(a)

print(a.shape)
```

```python
from numpy import array
import numpy as np

x = array([0, 10, 20, 30])
a = x[:, np.newaxis]
b = array([1, 2, 3])

print(a)
print()
print(b)
print()
print(a + b)
print()
print(a * b)
```

```python
import numpy as np

x = np.arange(56).reshape(7, 8)
print(x)

# 첨자 [1, 2]과 [3, 2]에 해당하는 값 10과 26
print(x[[1, 3], [2, 2]])

# 위와 동일
print(x[[1, 3], [2]])

# 1, 3행의 2열
print(x[[1, 3], 2])
```
