##다양한 데이터 프레임 생성

```python
import pandas as pd

# 방법 1
courses = pd.Series(["Spark", "PySpark", "Hadoop"], name="courses")
fees = pd.Series([250000, 300000, 200000], name="fees")

df1 = pd.DataFrame([courses, fees])
print(df1)
print()

print(df1.T)
print()

# 방법 2
df2 = pd.DataFrame()

df2[courses.name] = courses
df2[fees.name] = fees

print(df2)
```

```python
#여러 dict의 목록 데이터 사용
import pandas as pd

df1 = pd.DataFrame([
    {'국어': 70, '수학': 89, '영어': 90},
    {'국어': 80, '영어': 99},
    {'수학': 66, '영어': 78}
])
print(df1)
print()

s1 = pd.Series([90, 80, 60, 66])
s2 = pd.Series([45, 73, 83, 99])
s3 = pd.Series([92, 75, 82, 75])

df = pd.DataFrame({'국어': s1, '영어': s2, '수학': s3})
print(df)
print()

df.index = ['노', '장', '박', '강']
print(df)
```

```python
#데이터프레임의 결측값 표시 NaN
import pandas as pd

df1 = pd.DataFrame({
    'A': pd.Series([10, 20, 30], index=[0, 1, 2]),
    'B': pd.Series([3, 2, 1], index=[1, 2, 3])
})
print(df1)
print()

s1 = pd.Series([90, 80, 60], index=['노', '장', '박'])
s2 = pd.Series([45, 73, 99], index=['장', '박', '강'])
s3 = pd.Series([92, 75, 82, 75])

df2 = pd.DataFrame({'국어': s1, '영어': s2, '수학': s3},
                   index=['노', '장', '박', '강'])
print(df2)
```

```python
#df.dtypes
import pandas as pd
import numpy as np

d = {'col1': [1, 2], 'col2': [3, 4]}
df = pd.DataFrame(data=d)
print(df)
print()

print(df.dtypes)
print()

df = pd.DataFrame(data=d, dtype=np.int8)
print(df.dtypes)
```

##DataFrame의 다양한 참조

```python
#열 참조
import pandas as pd

pf = pd.DataFrame(
    data=[[25, 35, 8, 18],
          [18, 27, 10, 20],
          [17, 17, 10, 19],
          [12, 22, 9, 20],
          [22, 34, 8, 16]],
    index=['윤일형', '강수희', '홍소희', '유한빈', '신수빈'],
    columns=['중간', '기말', '과제', '출석']
)

print(pf)
print()

print(pf['중간'])
print()

print(pf.중간)
print()

print(pf['출석'])
print()

print(pf[['출석']])
print()

print(pf[['기말', '출석', '중간']])
print()

print(pf[['기말']])
```

```python
#행 참조
import pandas as pd

pf = pd.DataFrame(
    data=[[25, 35, 8, 18],
          [18, 27, 10, 20],
          [17, 17, 10, 19],
          [12, 22, 9, 20],
          [22, 34, 8, 16]],
    index=['윤일형', '강수희', '홍소희', '유한빈', '신수빈'],
    columns=['중간', '기말', '과제', '출석']
)

print(pf)
print()

print(pf['중간'] > 20)
print()

print(pf[pf['중간'] > 20])
print()

print(pf[pf['출석'] >= 19])
print()

print(pf[(pf['출석'] >= 19) & (pf['기말'] >= 20)])
```

```python
#loc
import pandas as pd

pf = pd.DataFrame(
    data=[[25, 35, 8, 18],
          [18, 27, 10, 20],
          [17, 17, 10, 19],
          [12, 22, 9, 20],
          [22, 34, 8, 16]],
    index=['윤일형', '강수희', '홍소희', '유한빈', '신수빈'],
    columns=['중간', '기말', '과제', '출석']
)

print(pf)
print()

print(pf.loc['윤일형'])
print()

print(pf.loc[['윤일형']])
print()

print(pf.loc[['홍소희', '윤일형']])
print()

print(pf.loc['윤일형':'유한빈'])
print()

print(pf.loc[pf.중간 > 20])
print()

print(pf.loc[(pf.중간 > 10) & (pf.과제 == 10)])
```

```python
#index 행 참조
import pandas as pd

pf = pd.DataFrame(
    data=[[25, 35, 8, 18],
          [18, 27, 10, 20],
          [17, 17, 10, 19],
          [12, 22, 9, 20],
          [22, 34, 8, 16]],
    index=['윤일형', '강수희', '홍소희', '유한빈', '신수빈'],
    columns=['중간', '기말', '과제', '출석']
)

print(pf)
print()

print(pf.loc[pf.index[0]])
print()

print(pf.loc[[pf.index[0]]])
print()

print(pf.loc[pf.index[1:4]])
print()

print(pf.loc[pf.index[1:4]][['기말', '과제']])
```

```python
#loc columns 열 참조
import pandas as pd

pf = pd.DataFrame(
    data=[[25, 35, 8, 18],
          [18, 27, 10, 20],
          [17, 17, 10, 19],
          [12, 22, 9, 20],
          [22, 34, 8, 16]],
    index=['윤일형', '강수희', '홍소희', '유한빈', '신수빈'],
    columns=['중간', '기말', '과제', '출석']
)

print(pf)
print()

print(pf.loc[:, pf.columns[0]])
print()

print(pf.loc[:, [pf.columns[0]]])
print()

print(pf.loc[:, pf.columns[1:]])
print()

print(pf.loc[:, pf.columns[1:]][['기말', '과제']])
```

```python
#df.iloc[정수] 참조
import pandas as pd

pf = pd.DataFrame(
    data=[[25, 35, 8, 18],
          [18, 27, 10, 20],
          [17, 17, 10, 19],
          [12, 22, 9, 20],
          [22, 34, 8, 16]],
    index=['윤일형', '강수희', '홍소희', '유한빈', '신수빈'],
    columns=['중간', '기말', '과제', '출석']
)

print(pf)
print()

print(pf.iloc[1])
print()

print(pf.iloc[[1]])
print()

print(pf.iloc[[1, 3]])
print()

print(pf.iloc[2:])
```

```python
#df.iloc[:, 정수] 참조
import pandas as pd

pf = pd.DataFrame(
    data=[[25, 35, 8, 18],
          [18, 27, 10, 20],
          [17, 17, 10, 19],
          [12, 22, 9, 20],
          [22, 34, 8, 16]],
    index=['윤일형', '강수희', '홍소희', '유한빈', '신수빈'],
    columns=['중간', '기말', '과제', '출석']
)

print(pf)
print()

print(pf.iloc[:, 2])
print()

print(pf.iloc[:, [2]]) #앞 뒤 둘중에 리스트로 한 번더 덮으면 데이터 프레임
print()

print(pf.iloc[:, 1:])
```

```python
#df.iloc[행번호, 열번호] 참조

import pandas as pd

pf = pd.DataFrame(
    data=[[25, 35, 8, 18],
          [18, 27, 10, 20],
          [17, 17, 10, 19],
          [12, 22, 9, 20],
          [22, 34, 8, 16]],
    index=['윤일형', '강수희', '홍소희', '유한빈', '신수빈'],
    columns=['중간', '기말', '과제', '출석']
)

print(pf)
print()

print(pf.iloc[1, 1])
print()

print(pf.iloc[1:3, 1:])
print()

print(pf.iloc[:, 1:])
print()

print(pf.iloc[[1, 3, 4], 1:])
print()

print(pf.iloc[[1, 3, 4], [0, 1]])
```

```python
#at['행명', '열명'] 참조 쓰는 이유는 이렇게 쓰면 빠르다고 해서 그렇다함...
import pandas as pd

pf = pd.DataFrame(
    data=[[25, 35, 8, 18],
          [18, 27, 10, 20],
          [17, 17, 10, 19],
          [12, 22, 9, 20],
          [22, 34, 8, 16]],
    index=['윤일형', '강수희', '홍소희', '유한빈', '신수빈'],
    columns=['중간', '기말', '과제', '출석']
)

print(pf)
print()

print(pf.at['윤일형', '과제'])
print()

print(pf.at['유한빈', '기말'])
print()

print(pf.at[pf.index[1], '출석'])
print()

print(pf.at[pf.index[3], pf.columns[2]])
```
