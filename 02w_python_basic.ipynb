```python
point = 92

if 90 <= point:
    print('A')
elif 80 <= point:
    print('B')
elif 70 <= point:
    print('C')
elif 60 <= point:
    print('D')
else:
    print('F')
```

### match
(switch 구문과 비슷)


```python
value = "banana"

match value:
  case 'apple':
    result = "사과"
  case 'banana':
    result = '바나나'
print(result)

value2 = "mango"

match value2:
  case "apple":
    result2 = "사과"
  case "banana":
    result2 = "바나나"
  case _: #else과 비슷
    result2 = None
print(result2)
```

### while문


```python
i = 1
sum = 0
while i <= 10:
  sum += i
  i += 1
print(sum)

a = 1
sum2 = 0
while True:
  sum2 += a
  a += 1
  if a > 10:
    break
print(sum2)
```

### lambda
(내가 못하는거)

```python
import math as m
(lambda a, b : m.sqrt(a**2 + b**2))(3, 4) #sqrt = 제곱근임 예를 들면 sqrt9면 = 3 나옴
```

```python
sum = (lambda a, b : a + b)
print(sum(10, 20))

nums = [1, 2, 3, 4, 5]
result = list(map(lambda x: x * 2, nums))
print(result)
```
